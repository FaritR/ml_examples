Избавимся сразу от переменных, названия которых заканчиваются на `_lognorm` (нужны для соревнования на Kaggle). Выберем их с помощью `filter()` и удалим `drop`-ом:

```python
df.drop(filter(lambda c: c.endswith('_lognorm'), df.columns), 
        axis = 1,       # axis = 1: столбцы 
        inplace = True) # избавляет от необходимости сохранять датасет
```

Столбец **`published`** (время публикации) содержит строки. Чтобы мы могли работать с этими данными как с датой/временем публикации, приведём их к типу `datetime`:

```python
print(df.published.dtype)
df['published'] = pd.to_datetime(df.published, yearfirst = True)
print(df.published.dtype)
```

Создадим несколько столбцов на основе данных о времени публикации:

```python
df['year'] = [d.year for d in df.published]
df['month'] = [d.month for d in df.published]
df['dayofweek'] = [d.isoweekday() for d in df.published]
df['day'] = [d.day for d in df.published]
df['hour'] = [d.hour for d in df.published]
```

## 3\. Когда лучше всего публиковать статью?

* Больше всего просмотров набирают статьи, опубликованные в 12 часов дня;
* У опубликованных в 10 утра постов больше всего комментариев;
* Больше всего просмотров набирают статьи, опубликованные в 6 часов утра;
* Максимальное число комментариев на гиктаймсе набрала статья, опубликованная в 9 часов вечера;
* На хабре дневные статьи комментируют чаще, чем вечерние.

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly
import plotly.graph_objs as go

init_notebook_mode(connected=True)
view_data = df.groupby('hour')[['views']].mean().join(df.groupby('hour')[['comments']].mean())
habr = df[df['domain'] == 'habrahabr.ru'][['hour'] + ['comments']]
geek = df[df['domain'] == 'geektimes.ru'][['hour'] + ['comments']]
habr_data = habr.groupby('hour')[['comments']].mean()
geek_data = geek.groupby('hour')[['comments']].mean()
max_data = geek.groupby('hour')[['comments']].max()
trace0 = go.Scatter(
    x = view_data.index,
    y = view_data.views,
    name = 'views')
trace1 = go.Scatter(
    x = view_data.index,
    y = view_data.comments,
    name = 'comments')
trace2 = go.Scatter(
    x = habr_data.index,
    y = habr_data.comments,
    name = 'habrahabr comments')
trace3 = go.Scatter(
    x = geek_data.index,
    y = geek_data.comments,
    name = 'geektimes comments')
trace4  = go.Scatter(
    x = max_data.index,
    y = max_data.comments,
    name = 'max geektimes comments')
data1=[trace0]
data2=[trace1, trace2, trace3]
data3=[trace4]
layout1 = {'title': 'Statistics of Views'}
layout2 = {'title': 'Statistics of Comments'}
layout3 = {'title': 'Max geek comments'}
fig1 = go.Figure(data=data1, layout=layout1)
fig2 = go.Figure(data=data2, layout=layout2)
iplot(fig1, show_link=False)
iplot(fig2, show_link=False)
fig3 = go.Figure(data=data3, layout=layout3)
iplot(fig3, show_link=False)
```
